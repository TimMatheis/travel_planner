```python
import pandas as pd
```

#### Load data

```python
# load csv
test_set = pd.read_csv("../data/destinations_v5.csv", sep=";")
```

```python
test_set
```

```python
season_dict = {
    'winter': ["Dec", "Jan", "Feb"],
    'spring': ["Mar", "Apr", "May"],
    'summer': ["Jun", "Jul", "Aug"],
    'autumn': ["Sep", "Oct", "Nov"]
}
```

#### Hard filter

```python
travel_categories = test_set["Category"].unique()

user_activity = input( "Choose one of the travel categories:" + travel_categories)

user_season = input( "Which season?").lower()
user_months = season_dict[user_season]
```

```python
test_set = test_set[test_set["Month"].isin(user_months)]
test_set
```

```python
test_set_clean = test_set.groupby(["City", "Category", "Category 2", "Budget"]).mean()
test_set_clean
```

```python
test_set_clean = test_set_clean.reset_index()
test_set_clean
```

```python
# filter for user category in both category columns
# watch out, user has to pass 'user_category' first!
test_set_clean = test_set_clean[(test_set_clean["Category"] == user_activity) | (test_set_clean["Category 2"] == user_activity)]
test_set_clean = test_set_clean.reset_index()
test_set_clean
```

#### User parameters questionnaire (input)

```python
preferences_dict = {
    1: {"category": "Budget", 
        "weight": 0},
    2: {"category": "Flight time", 
        "weight": 0},
    3: {"category": "CO2",
        "weight": 0},
    4: {"category": "Temperature",
        "weight": 0},
    5: {"category": "Rain probability",
        "weight": 0}
}
```

```python
i = 0
start_weight = 30
preferences_list = []
options_list = list( range(1,6) )
while i < 4:
    try: #the try statement runs the normal entry of 10 floats (question does not specify in float or integer, thus assume float)
        # print("Please state the number (e.g. 1) of your preference out of the remaining categories:\n")
        print(options_list)
        
        for count, option in enumerate(options_list):
            # print(count)
            print( "\t" + str(option) + "\t" + str(preferences_dict[options_list[count]]["category"]) + "\n")
            
        preference = int(input(f'Enter your preference {i+1} out of ' + str(options_list)))
        if (preference in options_list):
            options_list.remove(preference) 
            preferences_list.append(preference)
            preferences_dict[preference]["weight"] = start_weight
            start_weight -= 5
            i += 1 
            if len(preferences_list) == 4:
                preferences_dict[options_list[0]]["weight"] = start_weight
                preferences_list.append(options_list[0]) 
        else:
            print('You have to enter a number out of', options_list)

    except ValueError: #the except statement is triggered whenever a number is not entered
        print('You entered a string. Try again.')
print(f'\nYour preferences are: {preferences_dict[preferences_list[0]]["category"], preferences_dict[preferences_list[1]]["category"], preferences_dict[preferences_list[2]]["category"], preferences_dict[preferences_list[3]]["category"], preferences_dict[preferences_list[4]]["category"]}') #prints the final sum of all 10 entered integers
```

```python
preferences_dict
```

```python
# previous implementation (causes copy problems)
# df_scores = pd.DataFrame(columns=[test_set_clean.columns], index=test_set_clean.index).copy()
# df_scores[["City", "Category", "Category 2", "Budget"]] = test_set_clean[["City", "Category", "Category 2", "Budget"]].copy()

df_scores = test_set_clean.copy()
df_scores
```

```python
for x in ["CO2", "Flight time (min)", "Rain probability"]:
    for i in range(0, len(df_scores)):
        if test_set_clean.loc[i, x] <= test_set_clean[x].median():
            df_scores.loc[i, x] = 1
        else: 
            df_scores.loc[i, x] = 0

for y in ["Average Temp", "Sun hours per day"]:
    for j in range(0, len(df_scores)):
        if test_set_clean.loc[j, y] >= test_set_clean[y].median():
            df_scores.loc[j, y] = 1
        else: 
            df_scores.loc[j, y] = 0

replace_dic = {"Low": "1", "Medium": 0.5, "High": 0}
df_scores["Budget"].replace(replace_dic, inplace = True)

# previous implementation:
# df_scores['Budget'] = df_scores['Budget'].replace("Low", 1)
# df_scores['Budget'] = df_scores['Budget'].replace("Medium", 0.5)
# df_scores['Budget'] = df_scores['Budget'].replace("High", 0)

df_scores
```

```python
score_budget = preferences_dict[1]["weight"] * df_scores["Budget"]
score_flight_time = preferences_dict[2]["weight"] * df_scores["Flight time (min)"]
score_co2 = preferences_dict[3]["weight"] * df_scores["CO2"]
score_temp = preferences_dict[4]["weight"] * df_scores["Average Temp"]
score_rain = preferences_dict[5]["weight"] * df_scores["Rain probability"]

df_scores["Score"] = pd.concat([score_budget, score_flight_time, score_co2, score_temp, score_rain], axis=1).sum(axis=1)
df_scores
```

```python
# sort according to Score
df_scores.sort_values(by=['Score'], ascending = False, inplace = True)
df_scores["index"] = df_scores.reset_index().index
df_scores["index"] += 1
df_scores.rename({"index": "Rank"}, axis=1, inplace=True)
df_scores = df_scores.set_index("Rank")
df_scores
```
